### news NLP
* 기업 뉴스 크롤링을 통한 데이터 수집
* 수집된 자료 기반 BOW/TDM/TF-IDF를 통해 count 기반의 의미분석

```python
import sqlite3
import pandas as pd
import numpy as np
from konlpy.tag import * 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def getSub(whr = ''):
    dbname='navernews(2).db'
    with sqlite3.connect(dbname) as conn:
        cursor=conn.cursor()
        sql='select * from nnews ' + whr
        res=cursor.execute(sql).fetchall()
        return res
```

```python
res = getSub()
print(res)
```

```python
res[0]
```

```python
kkma = Kkma()
```

```python
fulldata = ' '.join([' '.join(r) for r in res])
fulldata[:100]
```

```python
t = fulldata.split(' ')   #t =terms 단어로 나눴다(split)
len(t)
```

```python
kpos = kkma.pos(fulldata) #전체 기사를 단어기반으로 나누고 전체 데이터 불러옴
```

```python
check = ['NNG','NNP','NNB','NNM','VV','VA','VXV','VXA','VCP','VCN']
```

```python
len(kpos)
```

```python
kpos[:5]
```

```python
bow={}
for k in kpos:
    if(k[1] in check):
        if(k[0] in bow.keys()):
            bow[k[0]]+=1
        else :
            bow[k[0]]=0
list(bow.keys())[:5] # word 확인
#len(bow.keys())  >> word count 확인(핵심단어)
```

```python
list(bow.values())[:10] #kpos에 나온 각 key값의 개수 확인
```

```python
bag = {}
for bk in list(bow.keys()):
    bag[bk]=0  # key 값을 빼서 초기화 했다는 뜻
```

```python
def makeBow(ws,b=bag):
    wbag = b.copy()
    kpos = kkma.pos(ws)
    for k in kpos:
        if(k[1] in check):
            wbag[k[0]]+=1
    return wbag
```

```python
resbag = makeBow(res[1][0])
list(resbag.values())
```

```python
TDMs = []
def makeTDM(s):
    resbag = makeBow(s)
    TDMs.append(list(resbag.values()))
```
