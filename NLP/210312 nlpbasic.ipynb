```python
fname='데이터/ys.txt'
with open(fname,'r',encoding='utf-8') as f : 
    res=f.readlines()
len(res)
```

```python
jres=' '.join(res)
#len(jres)
```

```python
res[0]
```

```python
p = kkma.pos(jres)
#p
```

```python
dictpos = {}
check=['NNG','NNP','NNB','NNM','VV','VA','VXV','VXA']
for w,pos in p:
    if (pos in check):
        if (w in dictpos.keys()):
            dictpos[w]+=1
        else :
            dictpos[w]=1
            #print('없음',w)
#dictpos
```

```python
len(list(dictpos))
```

```python
dictpos
```

```python
# top-down방식으로 많이 쓰는 단어로 내용을 추측할 수 있다.
sres=sorted(dictpos.items(),key=lambda x:x[1],reverse=True)
sres
# word count해서 단어의 가중치를 알아보겠다
```
