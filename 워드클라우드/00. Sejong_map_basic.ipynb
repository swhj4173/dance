### Folium 라이브러리를 통해 세종특별자치시 행정경계를 표현

```python
with open('./Data/32.세종시_행정경계(읍면동).geojson',mode='rt',encoding='utf-8') as f:
    geo = json.loads(f.read())
    f.close()

center = [36.599920, 127.298552]
m = folium.Map(location=center, zoom_start=11)

folium.Choropleth(
    geo_data=geo,
    fill_color='White',
).add_to(m)
m
```
