```python
import requests
from bs4 import BeautifulSoup
```

```python
job_result = requests.get("http://www.jobkorea.co.kr/Search/?stext=python")
```

```python
job_soup = BeautifulSoup(job_result.text,"html.parser")
```

```python
#post = job_soup.select('div >.post >.a')
```
