```python
def type_name(product):
    if 'EDP' in product:
        return 'EDP'
    elif 'EDT' in product:
        return 'EDT'
    elif 'EDC' in product :
        return 'EDC'
    elif 'PE' in product:
        return 'PE'
    elif None in product:
        return 'other'
    else:
        return 'other'
```
