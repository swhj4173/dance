```python
import sqlite3
import pandas as pd
dbname='perfume_all2.db'
with sqlite3.connect(dbname) as conn:
    cursor=conn.cursor()
    sql='select * from reviews'
    qry=cursor.execute(sql)
    desc=[row[0] for row in qry.description]
    data=qry.fetchall()
```

```python
pf=pd.DataFrame(data=data,columns=desc)
```

```python
pd.set_option('display.max_rows', 4000)
```

```python
apf=pf.groupby(['product'],as_index=False).agg(lambda x : x.mean() if x.dtype=='int64' else ' '.join(x))
apf['note']=pf['note']
apf['type']=pf['type']
apf['brand']=pf['brand']
```

```python
for i,v in enumerate(apf['product']):
    dups=apf[(apf['product'].str.contains(v))]['product']
    if len(dups)>1:    
        for dup in dups:
            print(i,dup)
```
